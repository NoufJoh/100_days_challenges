### Using matplotlib with Titanic dataset

```python
# stop showing warnings.
import warnings
warnings.filterwarnings("ignore")
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the Titanic data
titanic_data = pd.read_csv('titanic.csv')
```

```python
# Create a bar chart of the number of passengers who survived and did not survive, grouped by sex.
survival_data = titanic_data.groupby(['sex', 'survived']).size().unstack()
survival_data.plot(kind='bar', stacked=True)
plt.xlabel('Passenger Sex')
plt.ylabel('Number of Passengers')
plt.title('Passenger Survival by Sex')
plt.legend(['Did not survive', 'Survived'])
plt.show()
```
