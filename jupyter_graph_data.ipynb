```python
# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Load Data

```python
PATH_URL= "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"

# Save column names as a list
col_ls = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", 
         "race", "sex", "capital-gain", "capital-loss", "hours-per-week","native-country", "predicted_income"]

# Pass in col_ls into names parameter to explicitly name each column
df= pd.read_csv(PATH_URL, names=col_ls)
df.head()
```

```python
# describe all numerical features
df.describe()
```

```python
#  Look at the DataFrame shape
df.shape
```

## Graph 1:
This graph shows the relation between capital gain and capital loss. Clearly, there is no correlation.

```python
# replot showing relation
sns.relplot(x='capital-gain', y='capital-loss', data=df)

# Add title
plt.title("Capital Gain Vs. Capital Loss", size=16)

plt.show()
```

## Graph 2: 
This graph shows the Predicted Income Distribution. We can see the majority of the sample are expected to have an income of less than 50K with 75.92%.

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Predicted Income Distribution", size=16)

# define Seaborn color palette to use
palette_color = sns.color_palette('Set2')
data= df.groupby(['predicted_income']).size()
labels = data.keys()
# plotting data on chart
plt.pie(data, labels=labels, colors=palette_color, autopct='%.2f%%')

plt.show()
```

## Graph 3: 
This graph shows the age distribution. We can see that most of the ages are between 20 and 50.

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Age Distribution", size=16)

# histogram showing age distriution
sns.histplot(x=df['age'],data=df, alpha=0.6)

# Add label for vertical and horizantal axis
plt.xlabel("Age")
plt.show()
```
