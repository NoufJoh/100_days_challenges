```python
# stop showing warnings.
import warnings
warnings.filterwarnings("ignore")
# import panadas lib
import pandas as pd
```

```python
# Read the Titanic file
df = pd.read_csv("Titanic.csv")
# show head of data
df.head(5)
```

```python
# Print the number of rows and columns
print(df.shape)
# or
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Use the describe() method
print("Descriptive statistics:")
print(df.describe())
```

```python
# Use the info() method
print("Information about the data:")
df.info()
```

```python
# Print the sex, age, and fate (survived) fields
print("Sex, Age, and Survived fields:")
print(df[["sex", "age", "survived"]])
```

```python
# Determine male passengers
male_passengers = df[df["sex"] == "male"]
print(f"There are {len(male_passengers)} male passengers.")
```

```python
# Determine female passengers
female_passengers = df[df["sex"] == "female"]
print(f"There are {len(female_passengers)} female passengers.")
```
